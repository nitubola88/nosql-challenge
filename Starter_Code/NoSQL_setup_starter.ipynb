2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
query = {"BusinessType": "Restaurant/Cafe/Canteen"}
projection = {"BusinessTypeID": 1, "BusinessType": 1, "_id": 0}  # Include only BusinessTypeID and BusinessType, exclude _id

# Execute the query
results = establishments.find(query, projection)

# Display the result

for result in results:
    print(result)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
update_result = establishments.update_one(
    {"BusinessName": "Penang Flavours"},  # Query: Find the restaurant by BusinessName
    {"$set": {"BusinessTypeID": 1}}  # Update: Set the new BusinessTypeID (1 for Restaurant/Cafe/Canteen)
)
```

```python
updated_restaurant = establishments.find_one({"BusinessName": "Penang Flavours"})
# Confirm that the new restaurant was updated
# Display the updated information
from pprint import pprint
pprint(updated_restaurant)
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
dover_count = establishments.count_documents({"LocalAuthorityName": "Dover"})
print(f"Number of establishments in Dover: {dover_count}")
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
delete_result = establishments.delete_many({"LocalAuthorityName": "Dover"})

# Check how many documents were deleted
print(f"Number of establishments deleted: {delete_result.deleted_count}")
```

```python
# Check if any remaining documents include Dover
dover_count_after_delete = establishments.count_documents({"LocalAuthorityName": "Dover"})
print(f"Number of establishments in Dover after deletion: {dover_count_after_delete}")
```

```python
# Check that other documents remain with 'find_one'
query = {"LocalAuthorityName": {"$ne": "Dover"}}

# Find one document that doesn't have "Dover" as the LocalAuthorityName
remaining_document = establishments.find_one(query)
pprint(remaining_document)
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
update_result = establishments.update_many(
    {"geocode.latitude": {"$type": "string"}, "geocode.longitude": {"$type": "string"}},  # Find documents where latitude and longitude are stored as Decimal128
    [
        {"$set": {
            "geocode.latitude": {"$toDecimal": "$geocode.latitude"},  # Convert latitude to Decimal
            "geocode.longitude": {"$toDecimal": "$geocode.longitude"}  # Convert longitude to Decimal
        }}
    ]
)

# Print how many documents were modified
print(f"Updated {update_result.modified_count} documents with Decimal for latitude and longitude.")
#Updated 0 documents with Decimal for latitude and longitude.because it was already decimal
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
results=establishments.update_many({}, [ {'$set':{ "RatingValue" : {'$toInt': "$RatingValue"}}} ])
# Print how many documents were modified
print(f"Updated {results.modified_count} documents.")
```

```python
# Check that the coordinates and rating value are now numbers

updated_document = establishments.find_one({"BusinessName": "Oaks Nursing Home"})

# Check the data type of 'latitude', 'longitude', and 'RatingValue'
latitude_type = type(updated_document['geocode']['latitude'])
longitude_type = type(updated_document['geocode']['longitude'])
rating_value_type = type(updated_document['RatingValue'])

print(f"Latitude type: {latitude_type}")
print(f"Longitude type: {longitude_type}")
print(f"RatingValue type: {rating_value_type}")
```
